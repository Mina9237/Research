```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
matome_missing = pd.read_excel("./まとめ欠損地.xlsx")
matome_missing.head()
matome_missing.head()
```

```python
dates1 = pd.read_table("dates_kakunin.txt")
dates1.columns = ["Time"]
dates1["Time"] = dates1["Time"].astype(str)
dates1.dtypes
```

```python
dates1["Time"] = pd.to_datetime(dates1["Time"])
dates1.dtypes
dates1.head()
len(dates1)
```

```python
matome_missing2 = pd.concat([dates1,matome_missing])
matome_missing2.head()
```

```python
len(matome_missing2)
```

```python
matome_missing3 = matome_missing2[~matome_missing2["Time"].duplicated()]
```

```python
matome_missing3.head()
len(matome_missing3)
```

```python
matome_missing3
```

```python
matome_missing4 = matome_missing3.sort_values("Time").reset_index(drop =True)
matome_missing4["Time"].where(matome_missing4["Time"].value )
```

```python
matome_missing5 = matome_missing4.set_index("Time")
```

```python
matome_missing_6 = matome_missing5.interpolate(method='time')
matome_missing_6
```

```python
#matome_missing_6.to_excel('Chrolophyl_senkeihokan_time.xlsx')
```
